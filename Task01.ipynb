```python
import pandas as pd
import re
```

```python
# ==========================
# Load Dataset
# ==========================

df = pd.read_csv("/content/dirty_data_5000.csv")

print("\nOriginal Dataset")
print(df)
```

```python
# ==========================
# Remove Duplicate Records
# ==========================

df = df.drop_duplicates()

print("\nDuplicates Removed")
```

```python
# ==========================
# Handle Missing Values
# ==========================

# Fill missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Email
df['Email'] = df['Email'].fillna("unknown@gmail.com")

# Fill missing Join Date
df['Join_Date'] = df['Join_Date'].fillna("2025-01-01")
```

```python
# ==========================
# Standardize City Names
# ==========================

df['City'] = df['City'].str.strip().str.title()
```

```python
# ==========================
# Standardize Name Format
# ==========================

df['Name'] = df['Name'].str.strip().str.title()
```

```python
# ==========================
# Convert Date Formats
# ==========================

df['Join_Date'] = pd.to_datetime(
    df['Join_Date'],
    errors='coerce'
)

# Replace invalid dates
df['Join_Date'] = df['Join_Date'].fillna(
    pd.Timestamp("2025-01-01")
)
```

```python
# ==========================
# Validate Age
# ==========================

def fix_age(age):
    if age < 0 or age > 120:
        return df['Age'].median()
    return age

df['Age'] = df['Age'].apply(fix_age)
```

```python
# ==========================
# Validate Email
# ==========================

def validate_email(email):
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    if re.match(pattern, str(email)):
        return email
    return "invalid@email.com"

df['Email'] = df['Email'].apply(validate_email)
```

```python
# ==========================
# Structural Validation
# ==========================

expected_columns = [
    'Customer_ID',
    'Name',
    'Age',
    'Email',
    'Join_Date',
    'City',
    'Revenue'
]

if list(df.columns) == expected_columns:
    print("\nStructure Validation Passed")
else:
    print("\nStructure Validation Failed")
```

```python
# ==========================
# Data Quality Metrics
# ==========================

total_cells = df.size

missing_cells = df.isnull().sum().sum()

completeness = (
    (total_cells - missing_cells)
    / total_cells
) * 100

uniqueness = (
    len(df.drop_duplicates())
    / len(df)
) * 100

print("\nData Quality Report")
print("-------------------")
print(f"Completeness: {completeness:.2f}%")
print(f"Uniqueness: {uniqueness:.2f}%")
```

```python
# ==========================
# Save Clean Dataset
# ==========================

df.to_csv(
    "cleaned_data.csv",
    index=False
)

print("\nCleaned Dataset")
print(df)

print("\nFile Saved: cleaned_data.csv")
```
